```python
#Creating environment and importing necessory libraries
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

```

```python
# reading dataset and describing the dataset

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
women_degrees.head()
```

```python
#basic stat using pandas' describe
women_degrees.describe()
```

```python
# Generating a line chart that visualizes the
# historical percentage of Biology degrees awarded to women:

plt.plot(women_degrees['Year'],women_degrees['Biology'])
plt.xlabel('Women Degrees')
plt.ylabel('Biology')
plt.title('Biology degrees awarded to women')
plt.show()
```

#### From the plot, we can tell that Biology degrees increased steadily from 1970 and peaked in the early 2000's. We can also tell that the percentage has stayed above 50% since around 1987.
#### While it's helpful to visualize the trend of Biology degrees awarded to women, it only tells half the story. If we want the gender gap to be apparent and emphasized in the plot, we need a visual analogy to the difference in the percentages between the genders.


```python
#calculating number of degrees awarded to men in Biology
Biology_degrees_Awarded_to_men = 100 - women_degrees['Biology']
```

```python
Men_degrees = 100 - women_degrees['Biology']
plt.plot(women_degrees['Year'],women_degrees['Biology'],c='blue', label ='Women')
plt.plot(women_degrees['Year'],Men_degrees,c='green', label='Men')

plt.xlabel("Gender")
plt.ylabel('Year')
plt.title('Percentage of Biology Degrees Awarded By Gender')
plt.legend(loc='upper right')
plt.show()
```

#### The chart containing both line charts tells a more complete story than the one containing just the line chart that visualized just the women percentages. This plot instead tells the story of two distinct periods. In the first period, from 1970 to around 1987, women were a minority when it came to majoring in Biology while in the second period, from around 1987 to around 2012, women became a majority. You can see the point where women overtook men where the lines intersect.

```python
# Let's generate line charts for four STEM degree categories on a grid 

major_cats = ['Biology', 'Computer Science', 'Engineering', 'Math and Statistics']
fig = plt.figure(figsize=(12, 12))

for sp in range(0,4):
    ax = fig.add_subplot(2,2,sp+1)
    ax.tick_params(bottom='off',top='off',left='off',right='off')
    ax.spines["right"].set_visible(False)
    ax.spines["left"].set_visible(False)
    ax.spines["bottom"].set_visible(False)
    ax.spines["top"].set_visible(False)
    ax.plot(women_degrees['Year'], women_degrees[major_cats[sp]], c='blue', label='Women')
    ax.plot(women_degrees['Year'], 100-women_degrees[major_cats[sp]], c='green', label='Men')
    # Add your code here.
    ax.set_xlim(1968,2011)
    ax.set_ylim(0,100)
    ax.set_title(major_cats[sp])
# Calling pyplot.legend() here will add the legend to the last subplot that was created.
plt.legend(loc='upper right')
plt.show()
```

#### we can conclude that the gender gap in Computer Science and Engineering have big gender gaps while the gap in Biology and Math and Statistics is quite small. In addition, the first two degree categories are dominated by men while the latter degree categories are much more balanced.
